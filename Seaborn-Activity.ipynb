```python
%matplotlib inline

import pandas as pd
df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")
df.tail(10)
```

```python
import seaborn as sns
sns.set()
```

```python
#Scattterplot
sns.scatterplot(x="# Gears", y="CombMPG", data=df)
```

```python
#Boxplot
sns.set(rc={'figure.figsize':(15,5)})
a = sns.boxplot(x = '# Gears', y='CombMPG', data=df)
a.set_xticklabels(a.get_xticklabels(), rotation =45)
```

```python
#Distplot (Now known as Displot or Histplot)
sns.displot(x='# Gears', y='CombMPG', data=df)
```

```python
#Heatmap
import matplotlib.pyplot as plt
df2 = df.pivot_table(index='Mfr Name', columns='# Gears', values= 'CombMPG', aggfunc='mean')
sns.heatmap(df2)

#The color map on the left is focused on how much MPG a vehicle consumes for the gear it utilizes (over 718 vehicles are available in the given dataset)
```
