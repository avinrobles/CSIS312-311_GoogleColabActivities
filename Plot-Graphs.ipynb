```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x = np.arange(-2, 2, 0.01)
plt.plot(x, norm.pdf(x,0.5, 0.3))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x = np.arange(-2, 2, 0.01)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.fill_between(x, norm.pdf(x, 1.0, 0.5), color = "orange")
plt.show()
```

```python
plt.rcdefaults()

values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0.2, 0.01, 0.2, 0.1, 0.1]
labels = ['Dasma', 'Imus', 'Bacoor', 'Silang', 'Tagaytay']
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Student Locations')
plt.show()
```

```python
plt.rcdefaults()

values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
labels = ['Dasma', 'Imus', 'Bacoor', 'Silang', 'Tagaytay']
plt.bar(labels, values, color=colors)
plt.title("Highest Population in Provinces/Cities")
plt.show()
```
